## Task 1

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
```

```python
webpage = requests.get(link).text
```

```python
webpage = BeautifulSoup(webpage, "html.parser")
```

#### The data we are looking for is inside a table element wiith the class name "wikitable sortable plainrowheaders"

```python
table = webpage.find("table", class_="wikitable sortable plainrowheaders")
```

## Task 2

```python
th_tags = table.find_all('th', scope="row")
```

```python
state_names = []
for th in th_tags:
    a_tag = th.find('a')
    state_names.append(a_tag.string)
```

```python
state_names
```

## Task 3

```python
tr_tags = []
```

```python
for th in th_tags:
    tr_tags.append(th.parent)
```

```python
state_populations = []
```

```python
for tr in tr_tags:
    td_tags = tr.find_all("td")
    population_div = td_tags[-4].find("div")
    state_populations.append(population_div.string)
```

```python
state_populations
```

## Task 4

```python
state_data = pd.DataFrame({"state": state_names, "population": state_populations})
```

```python
state_data
```

```python
state_data.to_csv("us_info.csv")
```
